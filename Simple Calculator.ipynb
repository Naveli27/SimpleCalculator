```python
#simple calculator
import tkinter as tk
from tkinter import messagebox

def perform_calculation():
    num1 = float(entry_num1.get())
    num2 = float(entry_num2.get())
    operation = operation_var.get()

    if operation == "+":
        result = num1 + num2
    elif operation == "-":
        result = num1 - num2
    elif operation == "*":
        result = num1 * num2
    elif operation == "/":
        if num2 == 0:
            messagebox.showerror("Error", "Cannot divide by zero")
            return
        result = num1 / num2
    else:
        messagebox.showerror("Error", "Invalid operation")
        return

    result_label.config(text=f"Result: {result}")

# Create the main window
root = tk.Tk()
root.title("Simple Calculator")

# Entry widgets for numbers
entry_num1 = tk.Entry(root)
entry_num2 = tk.Entry(root)

# Radio buttons for operation choice
operation_var = tk.StringVar(value="+")  # Default operation is addition
add_button = tk.Radiobutton(root, text="+", variable=operation_var, value="+")
subtract_button = tk.Radiobutton(root, text="-", variable=operation_var, value="-")
multiply_button = tk.Radiobutton(root, text="*", variable=operation_var, value="*")
divide_button = tk.Radiobutton(root, text="/", variable=operation_var, value="/")

# Label to display the result
result_label = tk.Label(root, text="Result:")

# Button to perform calculation
calculate_button = tk.Button(root, text="Calculate", command=perform_calculation)

# Arrange widgets using grid layout
entry_num1.grid(row=0, column=0, padx=10, pady=10)
add_button.grid(row=0, column=1, padx=5, pady=10)
subtract_button.grid(row=0, column=2, padx=5, pady=10)
multiply_button.grid(row=0, column=3, padx=5, pady=10)
divide_button.grid(row=0, column=4, padx=5, pady=10)
entry_num2.grid(row=0, column=5, padx=10, pady=10)
calculate_button.grid(row=1, columnspan=6, padx=10, pady=10)
result_label.grid(row=2, columnspan=6, padx=10, pady=10)

# Start the GUI event loop
root.mainloop()
```
